```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

# pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherris","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherris","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherris","Dates"]
myexplode=[0.2,0,0,0]
mycolors=["Black","Blue","yellow","Green"]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolors)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
mycolors=["Black","Blue","yellow","Green"]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title="Four Fruits",loc='lower right')
plt.show()
```
